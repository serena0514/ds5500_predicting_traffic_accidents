```python
import requests

# Your HERE API key
api_key = '8GjMDPFdBN3iM1iLiwJspVekryXPVSan'

# Set the parameters for the request
base_url = 'https://api.tomtom.com/traffic/services/4/flowSegmentData/absolute/10/json'
point = '52.41072,4.84239'  # Latitude and Longitude of the point
params = {
    'key': api_key,
    'point': point,
    'unit': 'kmph',  # Optional parameter for speed unit
    'thickness': 10,  # Optional parameter for segment thickness
    'openLr': 'false'  # Optional parameter to include OpenLR code
}

# Make the GET request
response = requests.get(base_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    print("Flow Segment Data:")
    print(f"Functional Road Class: {data['flowSegmentData']['frc']}")
    print(f"Current Speed: {data['flowSegmentData']['currentSpeed']} km/h")
    print(f"Free Flow Speed: {data['flowSegmentData']['freeFlowSpeed']} km/h")
    print(f"Current Travel Time: {data['flowSegmentData']['currentTravelTime']} seconds")
    print(f"Free Flow Travel Time: {data['flowSegmentData']['freeFlowTravelTime']} seconds")
    print(f"Confidence: {data['flowSegmentData']['confidence']}")
    print(f"Road Closure: {data['flowSegmentData']['roadClosure']}")
    print("Coordinates:")
    for coordinate in data['flowSegmentData']['coordinates']['coordinate']:
        print(f" - Latitude: {coordinate['latitude']}, Longitude: {coordinate['longitude']}")
else:
    # Handle errors
    print(f"Failed to retrieve data: {response.status_code} - {response.text}")
```
